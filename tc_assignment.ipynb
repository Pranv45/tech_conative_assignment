Write a function to get distance 

```python
import numpy as np

def get_distance(lat_pickup, lon_pickup, lat_dropoff, lon_dropoff):
    R = 6371  # Radius of the earth in km
    dLat = deg2rad(lat_dropoff - lat_pickup)
    dLon = deg2rad(lon_dropoff - lon_pickup)
    a = (np.sin(dLat / 2) * np.sin(dLat / 2) + np.cos(deg2rad(lat_pickup)) * np.cos(deg2rad(lat_dropoff)) *np.sin(dLon / 2) * np.sin(dLon / 2))
    c = 2 * np.atan2(np.sqrt(a), np.sqrt(1 - a))
    d = R * c  # Distance in km
    return d

def deg2rad(deg):
    return deg * (np.pi / 180)
```

```python
#Get distances 
lat_pickup = data["pickup_latitude"]
lat_dropoff = data["dropoff_latitude"]
lon_pickup = data["pickup_longitude"]
lon_dropoff = data["dropoff_longitude"]
dist = get_distance(lat_pickup, lon_pickup, lat_dropoff, lon_dropoff)
```

```python
data = data.drop(["pickup_latitude", "pickup_longitude", "dropoff_latitude", "dropoff_longitude"], axis=1)
data["distance"] = dist
data
```
